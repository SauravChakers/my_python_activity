## Pre-processing textual data with Python: Normalisation and noise removal

You will usually need to clean and prepare the data before analysing it. Fortunately, Python offers useful libraries to automate this process. In addition to tokenisation, lets look at two more pre-processing actions: normalisation and noise removal. 

## Pre-processing demonstration: Normalisation

Normalisation aims to change the textual data into a comparable and uniform format. For example, all characters should be lowercase, remove numbers or convert to textual representation, remove punctuation, strip white space, and remove general English stop words. Normalisation processes are based on a detailed and nuanced understanding of grammatical rules and norms. 

### Remove stop words

```python
# Import all necessary libraries.
import nltk

# Download the stop words.
nltk.download('stopwords')

# Import the package.
from nltk.corpus import stopwords

# Create a set of English stop words.
stop_words = set(stopwords.words('english'))

# Display the set.
print(stop_words)
```

```python
text = """We took this ball to the beach and after close to 2 hours to pump it up, we pushed it around for about 10 fun filled minutes. That was when the wind picked it up and sent it huddling down the beach at about 40 knots. It destroyed everything in its path. Children screamed in terror at the giant inflatable monster that crushed their sand castles. Grown men were knocked down trying to save their families. The faster we chased it, the faster it rolled. It was like it was mocking us. Eventually, we had to stop running after it because its path of injury and destruction was going to cost us a fortune in legal fees. Rumor has it that it can still be seen stalking innocent families on the Florida panhandle. We lost it in South Carolina, so there is something to be said about its durability."""
```

```python
# Create an empty list for the filtered words.
filtered_text = []

# Assign the raw text data to a variable.
text = """We took this ball to the beach and after close to 2 hours to pump it up, we pushed it around for about 10 fun filled minutes. That was when the wind picked it up and sent it huddling down the beach at about 40 knots. It destroyed everything in its path. Children screamed in terror at the giant inflatable monster that crushed their sand castles. Grown men were knocked down trying to save their families. The faster we chased it, the faster it rolled. It was like it was mocking us. Eventually, we had to stop running after it because its path of injury and destruction was going to cost us a fortune in legal fees. Rumor has it that it can still be seen stalking innocent families on the Florida panhandle. We lost it in South Carolina, so there is something to be said about its durability."""

# Tokenise the text data.
# Create a tokenised word list.
tokenized_word = nltk.word_tokenize(text)

# Filter the tokenised words.
for each_word in tokenized_word:
    if each_word not in stop_words:
        filtered_text.append(each_word)

# Display the filtered list.
print("Tokenised list without stop words: {}".format(filtered_text))
```

### Reduce words using stemming and lemmatisation

#### Stemming

```python
# Import the necessary class.
from nltk.stem.snowball import SnowballStemmer

# Download the resource.
nltk.download('wordnet')

# Create a stemming object.
snow_stem = SnowballStemmer(language='english')

# Create a list of test words.
words = ['easily', 'durability', 'longest', 'wishing', 'worthwhile',
         'fantasizing', 'off-putting']

# Apply the stemming process to each word.
for word in words:
    print(word + "--->" + snow_stem.stem(word))
```

The output shows each original word and its stemmed version. As you can see, the Snowball Stemmer has mixed accuracy and usefulness. For example, the stemmer reduces wishing to wish, which is correct but returns the stem of easily as easili instead of eas.

#### Lemmatisation

```python
# Import the necessary class.
import nltk
from nltk.stem.wordnet import WordNetLemmatizer

# Download the lemma corpus.
nltk.download('omw-1.4')

# Create an instance of the class.
lemmatizer = WordNetLemmatizer()

# Define a text string to test.
text = "I love it when he purchases items that the kids and grandkids can't wait to try out. it's a lot of fun but he easily and accidentally bowls the toddlers over with them, so be careful."

# Create an empty output list.
lemmatized_words_list = []

# Tokenise the string.
tokenized_word = word_tokenize(text)

# Apply lemmatisation to each tokenised word.
for each_word in tokenized_word:
    lem_word = lemmatizer.lemmatize(each_word)
    lemmatized_words_list.append(lem_word)

# Display the output list.
print("Lemmatized Words list {}".format(lemmatized_words_list))
```

Notice how all the plurals have been changed to singular (e.g. kids to kid and toddlers to toddler). However, unrecognised words (e.g. grandkids) are unchanged. A limitation of these pre-processing tools is that they are only as good as their underlying language corpora. 

### Pre-processing demonstration: Noise Removal

Noise removal is a task-specific process rather than a language-specific process. It can:

- remove text file headers and footers
- remove HTML, XML, markup, and metadata
- extract valuable data from other formats (e.g. JSON) or in databases.

Noise removal depends on how data is originally captured or wrapped. Pattern matching (i.e. checking a sequence of tokens for the presence of a pattern) can help speed up this process as it can take up to 80% of data preparation time. In a single block of text, we can strip the text of formatting by, for example, removing punctuation. 

```python
# Import the necessary module.
import re

# Define some text. 
text = "Perfect! Buying a second. Using it to make a hot air balloon for an escape room adventure. Event in Oct. will share photos."

# Filter out the specified punctuation.
no_punct = re.sub(r"[\.\?\!\,\:\;\"]", '', text)

# Display the filtered text.
print(no_punct)
```

Here we can see the output no longer includes any punctuation marks.

## 3.1.9 Using Python to extract sentiment scores from customer reviews

### Prepare workstation

```python
# Install vaderSentiment tool.
!pip install vaderSentiment
```

```python
# Import the necessary class.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Create an object from the class.
sia = SentimentIntensityAnalyzer()
```

### Define Input Data

```python
# Define the input data.
sentences = [
    "Excellent customer service. Also loved the amazing showroom where you can get a real feel of the top quality furniture and get inspiration for room designs.",
    "I found the sales person to be quite rude and stuck up... don't think she realised I know all the suppliers she uses. I got the sofa & chair elsewhere. I will not be coming back",
    "Thank you for finding replacement crystals for my chandelier.",
    "I have worked with the team in several houses. All work was carried out on time and to budget."]
```

### Apply sentiment Analysis

```python
# Apply sentiment analysis.
for sentence in sentences:
    score = sia.polarity_scores(sentence)['compound']
    print(f'The sentiment value of the sentence :"{sentence}" is : {score}')
    # If you insert this line of code, you will have a line break between the outputs.
    print("="*50)
```

### Calculate the output

```python
# Calculate the percentage of the output.
for sentence in sentences:
    print(f'For the sentence "{sentence}"')
    # Calculate the scores.
    polarity = sia.polarity_scores(sentence)
    pos = polarity['pos']
    neu = polarity['neu']
    neg = polarity['neg']
    
    # Display the scores.
    print(f'The percentage of positive sentiment in :"{sentence}" is : {round(pos*100,2)} %')
    print(f'The percentage of neutral sentiment in :"{sentence}" is : {round(neu*100,2)} %')
    print(f'The percentage of negative sentiment in :"{sentence}" is : {round(neg*100,2)} %')
    print("="*50)
```

## 3.2.6 Developing text models with NLP

### Prepare the workstation

```python
# Import the necessary library.
import nltk

# Download the existing movie reviews.
nltk.download('movie_reviews')
```

### Construct a list of documents

```python
# Import the necessary libraries.
from nltk.corpus import movie_reviews
import random

# Construct a nested list of documents.
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]

# Reorganise the list randomly.
random.shuffle(documents)
```

```python
# Create a list of files with negative reviews.
negative_fileids = movie_reviews.fileids('neg')

# Create a list of files with positive reviews.
positive_fileids = movie_reviews.fileids('pos')

# Display the list lengths.
print(len(negative_fileids), len(positive_fileids))
```

```python
# View the output.
print(movie_reviews.raw(fileids=positive_fileids[0]))
```

#### Can see that the files are split into sentences, and the text is set to all lowercase.

### Define a feature extractor function

This function will teach the classifier which aspects of the data it should pay attention to.

```python
# Create an object to contain the frequency distribution.
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())

# Create a list that contains the first 2000 words.
word_features = list(all_words)[:2000]

# Define a function to check if each word is in the set of features.
def document_features(document): 
    # Create a set of document words.
    document_words = set(document)
    # Create an empty dictionary of features.
    features = {}
    # Populate the dictionary.
    for word in word_features:
       # Specify whether each feature exists in the set of document words. 
       features['contains({})'.format(word)] = (word in document_words)
    # Return the completed dictionary.
    return features
```

Testing the function quickly. Generate a dictionary for the first review.

```python
# Generate a dictionary for the first review
test_result = document_features(documents[0][0])

for key in test_result:
    print(key, ' : ', test_result[key])
```

This is just a selection of the much larger actual output, but you can see that the function works. The resulting dictionary contains every feature and specifies whether it exists in the documents list.

### Train the classifier
Now that you have defined the feature extractor, you will use it to train a Naive Bayes classifier to predict the sentiments of new movie reviews and check the accuracy.


```python
# Create a list of feature sets based on the documents list.
featuresets = [(document_features(d), c) for (d,c) in documents]

# Assign items to the training and test sets.
train_set, test_set = featuresets[100:], featuresets[:100]

# Create a classifier object trained on items from the training set.
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Display the accuracy score in comparison with the test set.
print(nltk.classify.accuracy(classifier, test_set))
```

The example output above shows an accuracy of 77%. Astounding, considering this is achieved without even tweaking or fine-tuning any parameters! 

See which features the classifier found to be most informative. You can specify 5 as a parameter to see the top five features.

```python
classifier.show_most_informative_features(5)
```

### Identify named entities

A named entity is, in the simplest sense, a proper noun. Naming refers to identifying whether the noun is a persons name, a street address, or the name of an organisation. We can perform further useful operations once nouns are identified and tagged with these various categories. For example, we might want to visualise the subject or extract the focus of a comment.

To perform this task, well turn to spaCy. Although there are several libraries that work well for named entity recognition, spaCy outshines the others when it comes to visualising the named entities. First, youll need to install the spaCy library and load the supporting resource for the English language. You can use pip to install spaCy. 

```python
# Import library.
import spacy

# Load the English resource.
nlp = spacy.load('en_core_web_sm')

# Return the text snippet and its corresponding entity label in a list.
def generate_named_entities(comment):
    return [(ent.text.strip(), ent.label_) for ent in nlp(comment).ents]
```

```python
survey['named_entities'] = survey['COMMENTS'].apply(generate_named_entities)

survey.head()
```

### Visualise named entities

```python
from spacy import displacy

# Iterate through a selection of comments.
for i in range(750,1750):
    # Check whether the corresponding comment has a named entity.
    if survey['named_entities'][i]:
        # Highlight the entity in the comment.
        displacy.render(nlp(survey['COMMENTS'][i]), style='ent', jupyter=True)
```

Here, we visualise the named entities. You can see the individual words are highlighted in the comments. The highlights indicate location for words such as 4th St  and person for mentions of names such as Madison. Although (as we can see here), the name Madison is actually the name of a street, which can only be understood in context.

Use this to get a quick sense of the nature of the responses without having to read through them. Visualising the comments with named entity labels can help to identify key topics within the data. Understand that the labels can be mistaken. Even from a glance, it is possible to see that this first attempt has some errors. For example, references to street names like 21st are labelled ordinal numbers, and references to Madison Dr are labelled 'person'. These techniques are also crucial to converting text into formats suitable for classification/prediction using machine learning and deep learning techniques.

### Generate a document term matrix

Here well generate a document-term matrix (DTM) using the bag-of-words model. This is the simplest way of extracting features from written text. It works by converting text into a matrix of the occurrence of words within a document. A document, in this case, refers to a single survey response. Each response constitutes a document. Well now turn to scikit-learn and the CountVectorizer class to generate a DTM.

```python
# Import the necessary classes.
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

# Use a tokenizer object to remove unwanted elements.
token = RegexpTokenizer(r'[a-zA-Z0-9]+')

# Create a CountVectorizer object to process the comments.
cv = CountVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),
                     tokenizer = token.tokenize)

# Apply the transformation to the comment data.
cvs= cv.fit_transform(survey['COMMENTS'])
```

Here, well set the parameters. For this demonstration, well only look at the first 15 documents. 

```python
# Create a DataFrame.
dt = pd.DataFrame(cvs.todense()).iloc[:15]  

# Name the columns.
dt.columns = cv.get_feature_names()

# Transpose columns and headings.
document_term_matrix = dt.T

# Update the column names.
document_term_matrix.columns = ['Doc '+str(i) for i in range(1, 16)]

# Get the totals.
document_term_matrix['total_count'] = document_term_matrix.sum(axis=1)

# Identify the top 10 words 
document_term_matrix = document_term_matrix.sort_values(by ='total_count', 
                                                        ascending=False)[:10] 

# Display the results.
print(document_term_matrix.drop(columns=['total_count']).head(10))
```

Looking at the DTM, youll see that each row indicates a unique word that appeared across the responses, while the columns represent a unique document (the individual survey responses). Now lets look at which words were most frequently used across the responses. Create a horizontal bar plot based on the total_count data.

```python
document_term_matrix['total_count'].plot.barh(figsize=(12, 9), fontsize=16, colormap ='plasma')
```

Since we only looked at a small sample of the data, the words highlighted here are quite different from some earlier examples. Working with small samples can be useful, since it speeds up the time and resources needed for analysis. However, the insights we can extract are somewhat limited or might be biased if the sample isnt representative enough.
